# 複雑な処理の例（複数テーブルを連結、条件付きデータ取得）

```python
## データを取得
cur = conn.cursor(dictionary=True) #connの生成後、1度しかカーソルオブジェクトは作れない？
cur.execute(
    "SELECT \
     A2.LotNumber   as SenseLot,\
     A2.WaferNumber as SenseWafer,\
     A2.BacksideID  as SenseWaferID,\
     A1.BacksideID  as LidWaferID,\
     LEFT(A1.BacksideID,'4') as ShortLidLot,\
     A3.* \
     \
     FROM\
     FabInventory.RawWafers A1\
     inner join FabInventory.SenseWafers A2\
           on   A1.BondedTo = A2.Id \
     inner join ExperimentalProber.SolidusSummary A3\
           on   A2.LotNumber = A3.Lot and A2.WaferNumber = A3.Wafer\
     \
     WHERE Record = 'All'\
     and A1.DeviceOrLid = 'Lid'\
     -- and A2.LotNumber like '8343%'\
     and DTime between '2018-01-01' and NOW();"
    
    #SELECT         :取得する列を指定（*を使うと全部指定）
    #FROM           :使用するテーブルを指定
    #inner join(on~):テーブルを連結　onは連結条件
    #WHERE          :取得するデータの条件
    #--             :コメントアウト
)
print(cur.statement)


#list_one = cur.fetchone()#cur.fetchmany()
list_all = cur.fetchall()  # 全部よみこまないとcloseできない
cur.close()                # closeしないと別の接続できない
#print(list_all[0])


## DataFrame形式に変換
df_concat = pd.DataFrame(list_all)
print('shape:',df_concat.shape)
display(df_concat)
```
